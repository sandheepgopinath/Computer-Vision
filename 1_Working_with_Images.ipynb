```
# Creating a filter 

def filter(image,size,stride):
  rows,columns,layers=image.shape
  #Calculating the shapes of the output image
  column=int(1+(columns-size)/stride)
  row=int(1+(rows-size)/stride)

  #Creating a dummy output image with black pixels
  output=np.zeros((row,column,layers))
  
  # Applying the filter calculations
  for r in range(0,row):
    for c in range(0,column):
      for l in range(0,layers):
        #Croping a part of the image to process
          row_start=r*stride
          row_end=row_start+size
          column_start=c*stride
          column_end=column_start+size
         
          sliced=image[row_start:row_end, column_start:column_end,l]
          output[r,c,l]=int(sliced.mean())
          
  return output
```

```
out=filter(img,3,2)
out=out/255
plt.imshow(out)
```

```
io.imsave('cat.jpg',out)
```
