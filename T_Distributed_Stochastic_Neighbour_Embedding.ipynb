```
import pandas as pd
from keras.datasets import mnist
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
```

```
(x_train,y_train),(x_test,y_test)=mnist.load_data()
```

```
X=np.zeros((x_train.shape[0],784))
```

```
for i in range(x_train.shape[0]):
  X[i]=x_train[i].flatten()
```

```
Y=y_train
```

```
#Inorder to select a sample of the data using sample(), the data should be in the format of a pandas dataframe
X_df=pd.DataFrame(X)

Y_df=pd.DataFrame(Y)
```

```
#Taking a small percentage of the data to do the TSNE

X_tsne=X_df.sample(frac=0.1,random_state=10)
Y_tsne=Y_df.sample(frac=0.1,random_state=10)
```

```
#Applying the TSNE function
import time
start=time.time()

tsne=TSNE(n_iter=300)
tsne_result=tsne.fit_transform(X_tsne)

print('TSNE completed in {}'.format(time.time()-start))
```

```
#Visualizing the TSNE Result

fig=plt.figure(figsize=(10,10))
plt.scatter(x=tsne_result[:,0],y=tsne_result[:,1],c=Y_tsne,cmap=plt.cm.get_cmap('Paired'))
plt.colorbar(ticks=range(10))
```
